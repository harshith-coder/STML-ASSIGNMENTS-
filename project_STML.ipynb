```python
import cv2
import time
from deepface import DeepFace

# Initialize the camera
cam = cv2.VideoCapture(0)
time.sleep(1)

# Load pre-trained face detector
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

while True:
    # Capture frame from camera
    _, img = cam.read()
    img = cv2.resize(img, (800, 600))  # Resize for better performance

    # Convert to grayscale for face detection
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Loop through detected faces
    for (x, y, w, h) in faces:
        # Draw rectangle around each face
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # Analyze the face for emotion
        face_img = img[y:y + h, x:x + w]
        result = DeepFace.analyze(face_img, actions=['emotion'], enforce_detection=False)

        # Access the first item if result is a list
        if isinstance(result, list):
            result = result[0]

        # Get dominant emotion and its accuracy
        emotion = result['dominant_emotion']
        accuracy = result['emotion'][emotion]  # Get confidence for the dominant emotion

        # Display emotion label with accuracy on the frame
        cv2.putText(img, f"Emotion: {emotion} ({accuracy:.2f}%)", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

        # Print emotion and accuracy to the terminal
        print(f"Detected Emotion: {emotion}, Accuracy: {accuracy:.2f}%")

    # Show the frame
    cv2.imshow("Facial Expression Detection", img)

    # Break loop on 'q' key press
    key = cv2.waitKey(10)
    if key == ord("q"):
        break

# Release resources
cam.release()
cv2.destroyAllWindows()
```
